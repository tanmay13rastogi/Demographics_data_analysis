```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

file_path = "C:\\Users\\Dell\\Downloads\\Demographic Data.xlsx"
```

```python
df=pd.read_excel(file_path)
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df['MAX SALARY'].isnull()
```

```python
null_counts = df.isnull().sum()
columns_with_nulls = null_counts[null_counts > 0]
print("\nSummary of columns with null values:")
print(columns_with_nulls)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
#DATA CLEANING-REMOVING ROWS WITH IRRELEVANT COLUMNS
df[df['MAX SALARY']!='Prefer not to answer']
```

```python
df.tail()
```

```python
#DATA CLEANING- REMOVING $ TO SEGREGATE NUMBERS
df=df[df['MAX SALARY'].str.contains('\$', na=False)]
```

```python
df.head()
```

```python
df['MAX SALARY'] = df['MAX SALARY'].replace('[\$,]', '', regex=True).astype(int)
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='City', y='SATISFACTION SCORE', palette='viridis', ci=None)
plt.title('Average Satisfaction Score by City', fontsize=16)
plt.ylabel('Average Satisfaction Score', fontsize=14)
plt.xlabel('City', fontsize=14)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
ethnicity_counts = df['ETHNICITY'].value_counts()
plt.pie(ethnicity_counts, labels=ethnicity_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"),shadow=True)
plt.title('Ethnicity Distribution', fontsize=16)
plt.show()
```

```python
average_salary_by_gender = df.groupby('GENDER')['MAX SALARY'].mean()
```

```python
plt.figure(figsize=(10, 6))
c=["pink","blue"]
plt.bar(average_salary_by_gender.index, average_salary_by_gender.values, color=c)
plt.title('Average Max Salary by Gender', fontsize=16)
plt.ylabel('Average Max Salary ($)', fontsize=14)
plt.xlabel('Gender', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
avg_max_salary_by_position=df.groupby('JOB_TITLE')['MAX SALARY'].mean()
c=["r","b","g","pink","orange"]
plt.bar(avg_max_salary_by_position.index,avg_max_salary_by_position.values,color=c)

plt.ylabel('Maximum Salary')
plt.title('Average maximum salary by Job title')
plt.xticks(rotation=90)
```

```python
#AVERAGE AGE 
```

```python
avg_age_by_job_title=df.groupby('JOB_TITLE')['AGE'].mean()
```

```python
c=["r","b","g","pink","orange"]
plt.bar(avg_age_by_job_title.index,avg_age_by_job_title.values,color=c)
plt.xticks(rotation=90)
plt.ylabel('AVERAGE AGE')
plt.title('Average age by job')
```

```python
#AVERAGE MAXIMUM SALARY BY EMPLOYEMENT STATUS AND GENDER
```

```python

average_salary = df.groupby(['EMPLOYMENT', 'GENDER'])['MAX SALARY'].mean().unstack()
plt.figure(figsize=(12, 8))
average_salary.plot(kind='bar', ax=plt.gca())
plt.title('Average Maximum Salary by Employment Status and Gender', fontsize=16)
plt.ylabel('Average Maximum Salary ($)', fontsize=14)
plt.xlabel('Employment Status', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Gender', fontsize=12, title_fontsize='13')
plt.tight_layout()
plt.show()






```
